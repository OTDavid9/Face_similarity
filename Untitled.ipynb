```python
import os
import base64

# Ensure the output directory exists
os.makedirs("Images", exist_ok=True)

# Adjust column names
base64_column = "ImageBase64"
bvn_column = "BVN"

# Loop through rows
for _, row in df.iterrows():
    base64_str = row[base64_column]
    bvn = str(row[bvn_column])

    # Skip if image data is missing
    if not base64_str:
        print(f"Skipping BVN {bvn}: No image data")
        continue

    # Strip data prefix if present
    if "," in base64_str:
        base64_str = base64_str.split(",")[-1]

    try:
        # Decode base64 and write file
        image_data = base64.b64decode(base64_str)
        filename = f"Images/{bvn}.png"
        with open(filename, "wb") as f:
            f.write(image_data)
        # print(f"Saved: {filename}")
    except Exception as e:
        print(f"Error saving BVN {bvn}: {e}")


print("Done !!!")
```

```python
import base64
import cv2
import numpy as np
import pandas as pd
from mtcnn.mtcnn import MTCNN
from keras_facenet import FaceNet

# Load data
df = pd.read_csv("training_data.csv")

# Filter rows where base64 is not valid
df = df[df["ImageBase64"].apply(lambda x: isinstance(x, str))].copy()

# Initialize face detector and embedder
detector = MTCNN()
embedder = FaceNet()

def decode_base64_image(base64_str):
    """Convert base64 string to OpenCV image."""
    try:
        img_data = base64.b64decode(base64_str)
        np_arr = np.frombuffer(img_data, np.uint8)
        img = cv2.imdecode(np_arr, cv2.IMREAD_COLOR)
        return img
    except Exception as e:
        print(f"Decoding error: {e}")
        return None

def get_face_embedding(img):
    """Detect face and return embedding (first face only)."""
    try:
        # Detect face
        faces = detector.detect_faces(img)
        if not faces:
            return None

        # Extract bounding box
        x, y, w, h = faces[0]['box']
        x, y = max(0, x), max(0, y)
        face = img[y:y+h, x:x+w]

        # Resize to 160x160 for embedding model
        face = cv2.resize(face, (160, 160))
        face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)

        # Get embedding
        embedding = embedder.embeddings([face])[0]  # returns a list of embeddings
        return embedding.tolist()
    except Exception as e:
        print(f"Embedding error: {e}")
        return None

# Process all images
embeddings = []

for idx, row in df[0:10].iterrows():
    img = decode_base64_image(row["ImageBase64"])
    if img is None:
        embeddings.append(None)
        continue

    embedding = get_face_embedding(img)
    embeddings.append(embedding)

# Add to DataFrame and save
df["FaceEmbedding"] = embeddings
df.to_csv("output_with_faces.csv", index=False)

print("✅ Face detection and embeddings completed.")
print(df[["FaceEmbedding"]].head())
```
